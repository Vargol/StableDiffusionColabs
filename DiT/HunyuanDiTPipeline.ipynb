```python
import mediapy as media
import random
import sys
import torch
import gc
```

```python
import torch
from diffusers import HunyuanDiTPipeline

pipe = HunyuanDiTPipeline.from_pretrained("Tencent-Hunyuan/HunyuanDiT-v1.2-Diffusers", torch_dtype=torch.float16)
pipe.to("cuda")
```

```python
# You may also use English prompt as HunyuanDiT supports both English and Chinese
prompt = "A photograph of Emily Booth as Sappho, Sappho stands on a rocky cliff, with a grey stormy sky over a stormy sea in the background, she is wears a black sheer dress that the wind is blowing against her legs, and has a harp on the ground by her side, one hand on the harp. the a a few sea birds far in the background. cold color palette, vivid colors, detailed, 8k, 35mm photo, Kodachrome, Lomography, highly detailed"
negative_prompt = "painting, drawing, illustration, glitch, deformed, mutated, cross-eyed, ugly, disfigured"
#prompt = "一个宇航员在骑马"
image = pipe(prompt, negative_prompt=negative_prompt).images[0]

media.show_image(image)
```
